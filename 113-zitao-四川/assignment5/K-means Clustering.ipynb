```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lenna.png', 0)
print(img.shape)
rows, cols = img.shape[:]
print(rows)
print(cols)
data = img.reshape((rows*cols, 1))
data = np.float32(data)
criteria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
flags = cv2.KMEANS_RANDOM_CENTERS

compactness, labels, centers = cv2.kmeans(data, 4, None, criteria, 10, flags)
print(compactness) # the minimun compactness value
print(labels) # output integer array that stores the cluster indices for every sample
print(centers) # Output matrix of the cluster centers, one row per each cluster center

dst = labels.reshape((img.shape[0], img.shape[1]))
```

```python
plt.subplot(1,2,1)
plt.imshow(img, 'gray')
plt.title('source image')
plt.xticks([])
plt.yticks([])
plt.subplot(1,2,2)
plt.imshow(dst, 'gray')
plt.title('clustering image')
plt.xticks([])
plt.yticks([])
plt.show()
```
